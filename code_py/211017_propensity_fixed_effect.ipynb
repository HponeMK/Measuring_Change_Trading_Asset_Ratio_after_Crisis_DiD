```python
# Load data, remove rows with missing values, create a date type column
df = pd.read_csv("/Users/hpone/Desktop/NUS MSBA/DBA5101/Grp Prj 2/DiD_data.csv")

df['date_parsed'] =  pd.to_datetime(df['rssd9999'], format='%Y%m%d')
```

```python
# Add column for different periods
df['pre_2006'] = 0
df.loc[df['date_parsed'] < '2006-12-31','pre_2006'] = 1
```

```python
# Get average of mean trading ratio per bank
df_affect_2007= df[df['pre_2006']==1].groupby('rssd9001')['bhc_avgtradingratio'].mean().reset_index()
df_affect_2007 = df_affect_2007.ffill()
df_affect_2007.rename(columns={"bhc_avgtradingratio": "affect_2007"}, inplace=True)
```

```python
# Merge average of means to data, set index, Create feature of affected entities
df = df.merge(df_affect_2007, on='rssd9001', how='inner')
df['date_parsed'] =  pd.to_datetime(df['rssd9999'], format='%Y%m%d')
df = df.set_index(['rssd9001', 'date_parsed'])
#df['dfa_affected_2007'] = df['after_DFA_1'] * df['affect_2007']
```
