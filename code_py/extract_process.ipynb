## Pre-processing

```python
import pandas as pd,\
  numpy as np,\
  statsmodels.formula.api as smfa,\
  pymatch.Matcher as pymm
```

```python
df_input = pd.read_csv(
  './DiD_data.csv',
  header='infer',
  index_col=False
)
```

```python
selected_ratios = [
  'bhc_avgtradingratio',
  'dep_roa1',
  'dep_leverage',
  'dep_lnassets',
  'dep_creditrisk_total3',
  'dep_cir',
  'dep_depositratio',
  'dep_loans_REratio',
  'dep_liquidity'
]
```

```python
#df_sample['after_DFA_1'] = df_sample['after_DFA_1'].astype('str')
#df_sample['treat_3_b_avg'] = df_sample['treat_3_b_avg'].astype('str')
df_input['rssd9001'] = df_input['rssd9001'].astype('str')

df_input['date_parsed'] =  pd.to_datetime(df_input['rssd9999'], format='%Y%m%d')
```

```python
#df_sample.pivot_table(index='date_parsed', columns='rssd9001',values='dep_liquidity',aggfunc=np.count_nonzero)
```

```python
#df_sample.groupby(['treat_3_b_avg','date_parsed'])[selected_ratios].describe().transpose()
```

```python

df_min_trading_ratio = df_input.groupby(['rssd9001'])['bhc_avgtradingratio'].min().reset_index()



df_sample = df_input.merge(right=df_min_trading_ratio , on ='rssd9001')

df_sample.rename(columns={'bhc_avgtradingratio_y':'min_trading_ratio','bhc_avgtradingratio_x':'bhc_avgtradingratio'},inplace=True)
```

## Difference in Difference - selected bank

```python
regout = smfa.ols('bhc_avgtradingratio ~ after_DFA_1 + rssd9001 ', df_sample[df_sample['min_trading_ratio'] >0]  ).fit()
```

```python
regout.summary2()
```

## Propensity matching

```python
m = pymm.Matcher(
  test=df_sample[(df_sample['treat_3_b_avg'] == 1) & ( df_sample['rssd9001'].isin(['1070345','1039502'])  ) ], 
  control=df_sample[(df_sample['treat_3_b_avg'] == 0) & ( df_sample['rssd9001'].isin(['1070345','1039502'])  ) ], 
  yvar = 'treat_3_b_avg', 
  exclude=['bhc_avgtradingratio','rssd9999','date_parsed']
)
```

```python


m.fit_scores(balance=False, nmodels=10)
```

## Fixed effects

```python
import linearmodels as lm
```

```python
df_weights = pd.read_csv('./data_propensity_matched.csv',header='infer',index_col=False)
```

# New Section

```python
df_weights['date_parsed'] =  pd.to_datetime(df_weights['rssd9999'], format='%Y-%m-%d')
```

```python
#df_fixed_effects = pd.concat(objs=[df_sample,df_weights],keys=['rssd9001','rssd9999'],join='inner')
```

```python
df_weights = df_weights.set_index(['rssd9001', 'date_parsed'])
```

```python
selected_ratios = [
  'dep_roa1',
  'dep_leverage',
  'dep_lnassets',
  'dep_cir',
  'dep_depositratio',
  'dep_loans_REratio',
  'dep_liquidity',
#  'bhc_avgtradingratio',
#  'after_DFA_1',
#  'treat_3_b_avg'
]
```

```python
#df_panel_set = df_fixed_effects[(df_fixed_effects['min_trading_ratio'] >0) & ( df_fixed_effects['weight'] >= 0 ) ]
df_panel_set = df_weights[ df_weights['weight'] >= 0 ]

fixed_effects = lm.PanelOLS(
  dependent=df_panel_set['bhc_avgtradingratio'],
  exog=df_panel_set[selected_ratios],
  entity_effects = True,
  time_effects=True,
  weights=df_panel_set['weight'],
#  rank_check=False
#  check_rank=False
)
```

```python
result = fixed_effects.fit(
  cov_type = 'clustered',
  cluster_entity=True,
             # cluster_time=True
)
```
